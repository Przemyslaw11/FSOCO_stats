```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df = pd.read_csv("reparsed.csv")
df.describe(percentiles=[0.01,0.05,0.1,0.25,0.50,0.75,0.9,0.95,0.99])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

'''
MatplotlibDeprecationWarning: Auto-removal of grids by pcolor() and 
pcolormesh() is deprecated since 3.5 and will be removed two minor releases later; 
please call grid(False) first.
'''
plt.grid(False) 

plt.hist2d(df["normCenterX"], df["normCenterY"], bins=(150,150), cmap="jet")
```

```python
plot  = sns.displot(df, x="pixWidth", hue="classTitle", kind="kde", palette=["blue", "orange", "red", "yellow", "gray"])
plot.set(xlim=(0,200))


# sns.displot(df, x="pixHeight", kind="kde")
# plt.hist(df["pixHeight"], bins=100)
```

```python
plot  = sns.displot(df, x="pixHeight", hue="classTitle", kind="kde", palette=["blue", "orange", "red", "yellow", "gray"])
plot.set(xlim=(0,200))
```

```python
# sns.pairplot(df, y="normHeight", x="normWidth", bins=(200,200), hue="classTitle", palette=["blue", "orange", "brown", "yellow", "gray"], kind="hist")
g = sns.FacetGrid(df, col="classTitle", hue="classTitle", palette=["blue", "orange", "brown", "yellow", "gray"])
g.map(sns.scatterplot, "normWidth", "normHeight", alpha=.7, size=0.7)
g.add_legend()
```

```python
g = sns.lmplot(
    data=df,
    x="normWidth", y="normHeight", hue="classTitle",
    palette=["blue", "orange", "red", "yellow", "gray"],
    height=5
)
```

```python
sns.set(rc={'figure.figsize':(8,8)})
sns.countplot(data=df, x='classTitle', order=df.classTitle.value_counts().index, palette=["yellow", "blue", "orange", "brown", "gray"], linewidth = 0.5)
print(f'Total number of cones is: {df.classTitle.value_counts().sum()}')
```
